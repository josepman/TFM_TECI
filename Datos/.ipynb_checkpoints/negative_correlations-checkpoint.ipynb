```python
# Load connectivity matrixes
files = os.listdir(healthy_fmri_path)
k = 0
G = []
for j in files:
    if k<1:
        df = pd.read_csv(healthy_fmri_path+j, header=None, delim_whitespace=True)
        print(j)
        np.fill_diagonal(df.values,1)   # make sure diagonal = 0
        normalized_df = preprocessing.normalize(df, norm='max')  # normalize
        individuals = normalized_df
    df = pd.read_csv(healthy_fmri_path+j, header=None, delim_whitespace=True)
    np.fill_diagonal(df.values,1)
    aux = aux + df
    np.fill_diagonal(aux.values,1)   # make sure diagonal = 0
    k = k + 1
    
avg_df = aux/len(files)   # average matrix of groups subjects
np.fill_diagonal(normalized_df,1)
normalized_df = preprocessing.normalize(avg_df, norm='max')  # normalize 
group_avg = normalized_df
group_avg = np.matrix(group_avg)
G_avg = nx.from_numpy_matrix(group_avg)
G_avg.name = "Original" 
G.append(G_avg)
```

```python
# No negative values
group_avg_0 = np.matrix(np.copy(group_avg))
group_avg_0[group_avg_0 < 0] = 0
G_0 = nx.from_numpy_matrix(group_avg_0)
G_0.name = "Only_pos_values" 
G.append(G_0)

# Absolute values
group_avg_abs = np.matrix(np.copy(group_avg))      
group_avg_abs = abs(group_avg_abs)
G_abs = nx.from_numpy_matrix(group_avg_abs)
G_abs.name = "Abs_values" 
G.append(G_abs)

# only negative values
group_avg_neg = np.matrix(np.copy(group_avg))      
group_avg_neg[group_avg_neg > 0] = 0   
group_avg_neg = abs(group_avg_neg)
G_neg = nx.from_numpy_matrix(group_avg_neg)
G_neg.name = "Only_neg_values" 
G.append(G_neg)
```

```python
## INDIVIDUAL SUBJECTS TABLE
tabla1 = []
tabla2 = []

for i in range(0,len(G)):
    tabla1.append(len(nx.nodes(G[i])))
    tabla2.append(len(nx.edges(G[i])))

columnas=['Original','Only_pos_values','Abs_values','Only_neg_values']
ind = {0:'Nodes',1:'Edges'}
tabla= pd.DataFrame([tabla1,tabla2],columns=columnas)
tabla.rename(index=ind)
```

```python
# CARACHTERISTICS OF INDIVIDUAL-SUBJECTS GRAPH
Gcluster=[]
Gdegree=[]
Gbet=[]
Gclo=[]
Geff=[]
Gasor=[]
Gseg=[]
Gint=[]
Gpath=[]

for i in range(len(G)):
    print(i)
    aux_degree = 0
    for v in G[i].nodes():
        aux_degree= aux_degree + G[i].degree(v, weight='weight')
    Gcluster.append(nx.average_clustering(G[i], weight='weight'))
    Gdegree.append((aux_degree)/(len(G[i].nodes())))
    Gbet.append(st.mean(list(nx.betweenness_centrality(G[i], weight='weight').values())))
    Gclo.append(st.mean(list(nx.closeness_centrality(G[i]).values())))
    Geff.append(nx.global_efficiency(G[i]))
    Gpath.append(nx.average_shortest_path_length(G[i]))
    Gasor.append(nx.degree_assortativity_coefficient(G[i]))
    Gseg.append(Gcluster[i]/Gcluster[0]) #se considera el grafo e-r
    Gint.append(Gpath[i]/Gpath[0])#integraci√≥n con la eficiencia(en lugar del path)
```

```python
# SUMMARY TABLE OF INDIVIDUAL GRAPHS
columnas=['Original','Only_pos_values','Absolute_values','Only_neg_values']
ind = {0:'Clustering',1:'GradoMedio',2:'BetweennessCentrality',3:'ClosenessCentrality',4:'GlobalEfficiency',5:'AverageShortestPath',6:'Asortatividad',7:'Segregaci√≥n(SW)',8:'Integraci√≥n(SW)'}
tabla= pd.DataFrame([Gcluster, Gdegree, Gbet, Gclo, Geff, Gpath, Gasor, Gseg, Gint], columns = columnas)
tabla.rename(index=ind)
```

```python
## INDIVIDUAL SUBJECTS TABLE
tabla1 = []
tabla2 = []

for i in range(0,len(G)):
    tabla1.append(len(nx.nodes(G[i])))
    tabla2.append(len(nx.edges(G[i])))

columnas=['Original','Only_pos_values','Abs_values','Only_neg_values']
ind = {0:'Nodes',1:'Edges'}
tabla= pd.DataFrame([tabla1,tabla2],columns=columnas)
tabla.rename(index=ind)
```

```python
def average_shortest_path_length_for_all(G):

    if nx.is_connected(G):
        # Normal case, the graph is connected 
        average=nx.average_shortest_path_length(G)
    else:
        # Try to see if the graph is not connected because of isolated nodes
        #G.remove_nodes_from(list(nx.isolates(G)))
        #if nx.is_connected(G):
            # Compute the graph average path without isolated nodes
        A = max(nx.connected_component_subgraphs(G), key=len)
        average = nx.average_shortest_path_length(A)

    return average
```

```python
# CARACHTERISTICS OF INDIVIDUAL-SUBJECTS GRAPH
Gcluster=[]
Gdegree=[]
Gbet=[]
Gclo=[]
Geff=[]
Gasor=[]
Gseg=[]
Gint=[]
Gpath=[]

for i in range(len(G)):
    print(i)
    aux_degree = 0
    for v in G[i].nodes():
        aux_degree= aux_degree + G[i].degree(v)
    Gcluster.append(nx.average_clustering(G[i]))
    Gdegree.append((aux_degree)/(len(G[i].nodes())))
    Gbet.append(st.mean(list(nx.betweenness_centrality(G[i]).values())))
    Gclo.append(st.mean(list(nx.closeness_centrality(G[i]).values())))
    Geff.append(nx.global_efficiency(G[i]))
    Gpath.append(average_shortest_path_length_for_all(G[i]))
    Gasor.append(nx.degree_assortativity_coefficient(G[i]))
    Gseg.append(Gcluster[i]/Gcluster[0]) #se considera el grafo e-r
    if average_shortest_path_length_for_all(G[i])==0:
        Gint.append('NaN')
    else:
        Gint.append(Gpath[i]/Gpath[0])#integraci√≥n con la eficiencia(en lugar del path)
```

```python
# SUMMARY TABLE OF INDIVIDUAL GRAPHS
columnas=['Original','Only_pos_values','Absolute_values','Only_neg_values']
ind = {0:'Clustering',1:'GradoMedio',2:'BetweennessCentrality',3:'ClosenessCentrality',4:'GlobalEfficiency',5:'AverageShortestPath',6:'Asortatividad',7:'Segregaci√≥n(SW)',8:'Integraci√≥n(SW)'}
tabla= pd.DataFrame([Gcluster, Gdegree, Gbet, Gclo, Geff, Gpath, Gasor, Gseg, Gint], columns = columnas)
tabla.rename(index=ind)
```
