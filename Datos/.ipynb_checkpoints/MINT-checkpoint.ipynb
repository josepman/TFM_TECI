```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import community as cm #python-louvain
import statistics as st
import operator as op
import itertools
```

```python
n=110 # nodos para er
m=1200# aristas para er
Grafos=[]
Grandom=nx.gnm_random_graph(n,m,seed=123456789) #generador de grafo er, semilla fija
Gregular=nx.grid_2d_graph(10,11) #generador de grafo regular
Gstar=nx.star_graph(n-1) #generador de grafo en estrella

Grafos.append(Grandom)
Grafos.append(Gregular)
Grafos.append(Gstar)

#importamos cada uno de los sujetos para estudio estructural y funcional
lista_sujetos = ['fmrisano.txt','fmriautismo.txt','fmriADHD.txt','dtisano.txt','dtiautismo.txt','dtialz.txt']
for sujeto in lista_sujetos:
    raw_data = pd.read_csv(sujeto,header=None,delim_whitespace=True)
    if sujeto == 'dtialz.txt':
        thr = 0.001
    raw_data[np.absolute(raw_data) > thr] = 1 #threshold para formar conexiones
    raw_data[np.absolute(raw_data) < thr] = 0
    matrizAdy= np.matrix(raw_data)
    Gsujeto = nx.from_numpy_matrix(matrizAdy)
    Grafos.append(Gsujeto)
```

```python
#calculamos las métricas globales para visualizarlas
Gcluster=[]
Gdegree=[]
Gbet=[]
Gclo=[]
Geff=[]
Gasor=[]
Gseg=[]
Gint=[]
Gpath=[]
for i in range(len(Grafos)):
    aux_degree = 0
    for v in Grafos[i].nodes():
        aux_degree= aux_degree + Grafos[i].degree(v)
    Gcluster.append(nx.average_clustering(Grafos[i]))
    Gdegree.append((aux_degree)/(len(Grafos[i].nodes())))
    Gbet.append(st.mean(list(nx.betweenness_centrality(Grafos[i]).values())))
    Gclo.append(st.mean(list(nx.closeness_centrality(Grafos[i]).values())))
    Geff.append(nx.global_efficiency(Grafos[i]))
    Gpath.append(nx.average_shortest_path_length(Grafos[i]))
    Gasor.append(nx.degree_assortativity_coefficient(Grafos[i]))
    Gseg.append(Gcluster[i]/Gcluster[0]) #se considera el grafo e-r
    Gint.append(Gpath[i]/Gpath[0])#integración con la eficiencia(en lugar del path)
```

```python
#dataFrame de visualización global
columnas=['Erdos-Renyi','Regular-Graph','Star-Graph','FMRI-sano','FMRI-autismo','FMRI-ADHD','DTI-sano','DTI-autismo','DTI-Alzheimer']
ind = {0:'Clustering',1:'GradoMedio',2:'BetweennessCentrality',3:'ClosenessCentrality',4:'GlobalEfficiency',5:'AverageShortestPath',6:'Asortatividad',7:'Segregación(SW)',8:'Integración(SW)'}
tabla= pd.DataFrame([Gcluster,Gdegree,Gbet,Gclo,Geff,Gpath,Gasor,Gseg,Gint],columns=columnas)
tabla.rename(index=ind)
```

```python
Local_prox = []
Local_bet = []
Local_clo = []
Local_cluster = []
Grado_node = []
for i in range(0,len(Grafos)):
    Local_prox.append(nx.average_neighbor_degree(Grafos[i]))
    Local_bet.append(nx.betweenness_centrality(Grafos[i]))
    Local_clo.append(nx.closeness_centrality(Grafos[i]))
    Local_cluster.append(nx.clustering(Grafos[i]))
    grado={}
    for v in Grafos[i].nodes():
            grado[v]=Grafos[i].degree(v)
    Grado_node.append(grado)
```

```python
for i in range(0,len(Grafos)):
    Local_prox[i] = sorted(Local_prox[i].items(),reverse=True,key=op.itemgetter(1))
    Local_bet[i] = sorted(Local_bet[i].items(),reverse=True,key=op.itemgetter(1))
    Local_clo[i] = sorted(Local_clo[i].items(),reverse=True,key=op.itemgetter(1))
    Local_cluster[i] = sorted(Local_cluster[i].items(),reverse=True,key=op.itemgetter(1))
    Grado_node[i] = sorted(Grado_node[i].items(),reverse=True,key=op.itemgetter(1))
```
